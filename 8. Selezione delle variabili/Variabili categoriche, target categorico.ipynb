```python
import pandas as pd
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif, chi2
```

```python
from sklearn.preprocessing import KBinsDiscretizer, LabelEncoder
```

```python
df = pd.read_csv("sample_dataset.csv").iloc[:,[0,1,2,3,4,-1]].dropna()
```

```python
X = df.iloc[:,0:5]
y = df['target']
```

```python
discr = KBinsDiscretizer(n_bins = 10,encode='ordinal')
X2 = discr.fit_transform(X)
```

```python
le = LabelEncoder()
y2 = le.fit_transform(y)
```

# Mutual Information

```python
selector = SelectKBest(lambda X,y : mutual_info_classif(X,y,discrete_features=True), k = 3)
```

```python
selector.fit_transform(X2,y2)
```

# Chi quadro

```python
selector2 = SelectKBest(chi2, k = 3)
```

```python
selector2.fit_transform(X2,y2)
```
