```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
df = pd.read_csv("sample_dataset.csv").iloc[:,0:10].dropna()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
selector = SelectFromModel(RandomForestRegressor(), threshold="mean")
```

```python
selector.fit_transform(X,y)
```

```python
selector2 = SelectFromModel(RandomForestRegressor(), threshold="median")
selector2.fit_transform(X,y)
```

```python
selector3 = SelectFromModel(RandomForestRegressor(), threshold=0.1)
selector3.fit_transform(X,y)
```

```python
selector4 = SelectFromModel(RandomForestRegressor(), threshold=-np.inf, max_features=3)
selector4.fit_transform(X,y)
```
