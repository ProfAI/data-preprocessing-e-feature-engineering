```python
import pandas as pd

from sklearn.feature_selection import SelectKBest, f_regression, r_regression, mutual_info_regression
```

```python
df = pd.read_csv("sample_dataset.csv").iloc[:,0:11].dropna()
```

```python
X = df.iloc[:,0:10]
y = df.iloc[:,10]
```

# F-score

```python
selector = SelectKBest(f_regression,k = 5)
```

```python
selector.fit_transform(X,y)
```

```python
X.columns[selector.get_support()]
```

# Indice di Pearson

```python
selector2 = SelectKBest(r_regression, k = 3)
```

```python
selector2.fit_transform(X,y)
```

# Mutual information

```python
selector3 = SelectKBest(mutual_info_regression, k = 4)
```

```python
selector3.fit_transform(X,y)
```
