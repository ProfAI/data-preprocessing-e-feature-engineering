```python
import pandas as pd
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = pd.read_csv("sample_dataset.csv").iloc[:,0:5].dropna()
```

```python
df
```

# Binning con bin a larghezza fissa

```python
binner1 = KBinsDiscretizer(strategy='uniform', n_bins=10, encode="ordinal")
```

```python
binner1.fit_transform(df)
```

```python
binner2 = KBinsDiscretizer(strategy='uniform', n_bins=10, encode="onehot-dense")
```

```python
binner2.fit_transform(df)
```

```python
binner2.fit_transform(df).shape
```

# Binning basato sui quantili

```python
binner3 = KBinsDiscretizer(strategy='quantile',n_bins=10,encode='ordinal')
```

```python
binner3.fit_transform(df)
```

# Binning basato su KMeans

```python
binner4 = KBinsDiscretizer(strategy='kmeans',n_bins = 10, encode='ordinal')
```

```python
binner4.fit_transform(df)
```
